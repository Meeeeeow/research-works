```python
def bar_chart(feature):
  survived = train_data[train_data["Survived"] == 1][feature].value_counts()
  dead = train_data[train_data["Survived"] == 0][feature].value_counts()
  df = pd.DataFrame([survived, dead]);
  df.index = ['Survived', 'Dead'];
  df.plot(kind='bar', stacked = True, figsize = (10,5))
```

```python
fare_ranges = [ 0, 10, 20, 30, float('inf')]
fare_ranges
```

```python
fare_labels =[ '<10','10-20','20-30','30+']
```

```python
train_data["Fare_Category"] = pd.cut(train_data['Fare'], bins=fare_ranges, labels=fare_labels)
train_data["Fare_Category"]
```

```python
train_data["Fare_Category"].value_counts().sort_index()
```

```python
train_data
```

```python
# @title Survival Rate by Age Group

sns.histplot(x="Age", hue="Survived", data=train_data, bins=10)
```

```python
survival_counts = train_data.groupby(["Fare_Category", "Sex", "Pclass"])["Survived"].sum().reset_index(name='Survival_Counts')
survival_counts
```

```python
grouped_data = train_data.groupby(['Fare_Category', 'Sex', 'Pclass'])['Survived'].agg(['sum', 'count']).reset_index()
grouped_data
```

```python
grouped_data['Survival_Percentage'] = round((grouped_data['sum'] / grouped_data['count']) * 100,2)
grouped_data['Survival_Percentage']
```

```python
grouped_data = grouped_data.rename(columns={'sum': 'Survival_Counts', 'count': 'Total_Counts'})

grouped_data = grouped_data[grouped_data["Total_Counts"] != 0]
grouped_data
```

```python
grouped_percentage_data = grouped_data[grouped_data["Survival_Percentage"] >= 80.00]
grouped_percentage_data
```

```python
grouped_low_data = grouped_data[grouped_data["Survival_Percentage"] <= 80.00]
grouped_sex_data =grouped_low_data[grouped_low_data["Sex"] != "male"]
grouped_sex_data
```

```python
test_data["Fare_Category"] = pd.cut(test_data['Fare'], bins=fare_ranges, labels=fare_labels)
test_data
```

```python
categories_to_check = [
    ('10-20', 'female', 2),
    ('20-30', 'female', 1),
    ('20-30', 'female', 2),
    ('20-30', 'female', 3),
    ('30+', 'female', 3),
    ('30+', 'female', 1),
    ('30+', 'female', 2)
]
```

```python
for category in categories_to_check:
    fare_category, sex, pclass = category
    category_count = test_data[
        (test_data['Fare_Category'] == fare_category) &
        (test_data['Sex'] == sex) &
        (test_data['Pclass'] == pclass)
    ].shape[0]
    print(f"Count for the category '{fare_category}', {sex}, {pclass} in the test data:", category_count)
```

```python
pred_4 = test_data[["PassengerId"]].copy()
pred_4["Survived"] = 0
pred_4["Survived"] = np.where(test_data["Sex"] == "female" , 1, 0)
pred_4
```

```python
conditions = (
    (test_data["Sex"] == "female") &
    (test_data["Fare_Category"] == "30+") &
    (test_data["Pclass"] == 3)
)
pred_4.loc[conditions, "Survived"] = 0
```
